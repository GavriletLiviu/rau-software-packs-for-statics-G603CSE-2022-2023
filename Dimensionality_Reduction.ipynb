```python
from sklearn.datasets import load_wine

X, y = load_wine(return_X_y=True)
X.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

model = PCA(n_components=2)
model.fit(X)

X_transformed = model.transform(X)
```

```python
X_transformed.shape
```

```python
import matplotlib.pyplot as plt

_ = plt.figure(figsize=(10, 10))
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], s=100, alpha=0.5, c=y)
plt.show()
```

```python
model.explained_variance_ratio_
```

```python
pca_3d = PCA(n_components=3)
pca_3d.fit(X)
X_transformed_3d = pca_3d.transform(X)

_ = plt.figure(figsize=(10, 10))
plt.scatter(X_transformed_3d[:, 0], X_transformed_3d[:, 1], s=100, alpha=0.5, c=y)
plt.show()
```

```python
pca_3d.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

y_pred = tree.predict(X_test)
y_test = np.array(y_test)
accuracy_score(y_true=y_test, y_pred=y_pred)
```
